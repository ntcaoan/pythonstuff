## The basic machine learning workflow
When you are trying to develop a model using machine learning, go thru those workflows
    1. examine and understand the data
    2. build an input pipeline
    3. build the model
    4. train the model
    5. test the model
    6. improve the model and repeat the process
   

### Step 1: Understand the data
Images can be stored in lots of different format -> but we will use a library to handle reading them in, so we don't have to worry about how they are stored

### Step 5: Test the model
- Now we want to see how well our model is actually performing
- The training process will produce some metrics we can judge this by - most notably, accuracy and loss
- It produces those values for both the training data and the validation data
- The values for the training data are used to improve the model through *epochs*, so the accuracy should naturally increase and the loss decrease
- What we really are watching is our validation loss and accuracy
- Seeing accuracy increase means the model is getting better at correctly predicting
